# Get all the S or T position from the 137 protein sequences

```python
os.chdir("/homes/t326h379/One_hundred_thirty_seven")
```

```python
import glob

Fasta_files = []
for name in glob.glob('*.fasta'):
    Fasta_files.append(name)
```

```python
len(Fasta_files)
```

```python
from Bio import SeqIO

Positive_and_negative_sites = []
for value in Fasta_files:
    name_of_the_file = value
    for record in SeqIO.parse(name_of_the_file, "fasta"):
        Protein_name = record.id.split("|")[1]
        Sequence = str(record.seq)
        for i, amino_acid in enumerate(Sequence):
            if amino_acid == "S" or amino_acid == "T":
                Positive_and_negative_sites.append((Protein_name,i+1))
                
Positive_and_negative_sites = set(Positive_and_negative_sites)
```

```python
len(Positive_and_negative_sites)
```

# Negative "S/T" Sites from 137 Proteins

```python
Negative_Sites_only = Positive_and_negative_sites - Positive_sites_only
```
